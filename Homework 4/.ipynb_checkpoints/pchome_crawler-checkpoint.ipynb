## Crawler part

```python
import requests
import time
```

```python
## get and find data
all_datas=[]
num = 0
page=0

while num < 300:
    
    page=page+1
    print(page)
    url = f'https://ecshweb.pchome.com.tw/search/v3.3/all/results?q=%E6%89%8B%E6%A9%9F&page={str(page)}&sort=sale/dc'
    
    a = requests.get(url)
    result = a.json()
    
    all_datas.append(result['prods'])
    num += 20
    time.sleep(0.5)
```

```python
all_datas
```

```python
def flatten_list(_2d_list):
    flat_list = []
    # Iterate through the outer list
    for element in _2d_list:
        if type(element) is list:
            # If the element is of type list, iterate through the sublist
            for item in element:
                flat_list.append(item)
        else:
            flat_list.append(element)
    return flat_list
```

```python
new_datas = flatten_list(all_datas)
new_datas
```

```python
## clean data
for product in new_datas:
    del product['picS']
new_datas
```
